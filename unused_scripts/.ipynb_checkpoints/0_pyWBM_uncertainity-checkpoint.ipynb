## Uncertainity Testing

### Single Parameter Change

```python
import os
import glob
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

# ---------------------------------------------------------------------------
# Configuration section
# Here we define our "fixed" and "variable" parameters.
# You can easily switch whats fixed and whats iterated over just by changing
# these definitions.
# ---------------------------------------------------------------------------

# Fixed parameters
fixed_model = "GFDL-ESM4"
fixed_directory = "/storage/group/pches/default/users/dcl5300/wbm_soilM_uc_2024_DATA/projections/eCONUS/out/LOCA2"
fixed_start_year = 2030
fixed_end_year = 2035

target_lon = -97.916944
target_lat = 27.518611
location = 'King_Ranch_Texas'

# Variable parameters

possible_dirs = ["SMAP", "VIC", "NOAH", "MOSAIC"]
possible_dirs = ["ssp245", "ssp370"]

possible_dirs = [
    "kge", "mae", "mse", "nse", "outer20rmse", "outer20ubrmse", "outer50rmse",
    "outer50ubrmse", "rmse", "taylor", "ubmae", "ubsme", "ubrmse"]

base_pattern = "*r1i1p1f1_ssp370_MOSAIC_{}.nc"  # one placeholder for sub_dir

# If you want to vary something else (like models), you could define a list of models:
# possible_models = ["GFDL-ESM4", "AnotherModel", "YetAnotherModel"]
# and keep sub_dir fixed by removing the loop over possible_dirs
# and adding a loop over possible_models, etc.

# ---------------------------------------------------------------------------
# Data Processing and Plotting
# ---------------------------------------------------------------------------

plt.figure(figsize=(12, 4))
ci_range = []

for sub_dir in possible_dirs:
    pattern = base_pattern.format(sub_dir)
    files = glob.glob(os.path.join(fixed_directory, pattern))
    
    if not files:
        continue

    for f in files:
        model_name = os.path.basename(f).split("_")[0]

        if model_name == fixed_model:
            ds = xr.open_dataset(f)
            ds_small = ds.sel(time=slice(f"{fixed_start_year}-01-01", f"{fixed_end_year}-12-31"))

            subset = ds_small.sel(lon=target_lon, method='nearest').sel(lat=target_lat, method='nearest')
            ci_range.append(subset.soilMoist.values)

            subset.soilMoist.plot(alpha=0.05, color='black')

            ds.close()
            del ds_small

# Once we have gone through all files (for all sub_dirs for the fixed model)
# we can compute the uncertainty range
time_values = subset['time'].values
ci_95th = np.percentile(ci_range, 97.5, axis=0)
ci_5th = np.percentile(ci_range, 2.5, axis=0)

plt.fill_between(
    time_values, 
    ci_5th, 
    ci_95th, 
    color='b', 
    alpha=.1, 
    label=f'95% Confidence Interval {fixed_model} {base_pattern[1:-1]}'
)

plt.title(f"{fixed_start_year}-{fixed_end_year} Inter SSP Soil Moisture Uncertainty, {location}")
plt.xlabel("Time")
plt.ylabel("Soil Moisture")
plt.ylim(150, 450)
plt.legend(fontsize='small', loc='upper right')

# Uncomment to save the figure if desired
base_path = "/storage/home/cta5244/work/0_soilmoisture/0_uncertainity_figures/"
plt.savefig(base_path + (f'1_soil_moist_model_{fixed_model}_{base_pattern[1:-1]}_{fixed_start_year}_{fixed_end_year}_{location}.png').replace("{", "-").replace("}", "-"))

#plt.show()
```

### Trend in Soil Moisture

```python
import os
import glob
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

# ---------------------------------------------------------------------------
# Configuration section
# Here we define our "fixed" and "variable" parameters.
# You can easily switch whats fixed and whats iterated over just by changing
# these definitions.
# ---------------------------------------------------------------------------

# Fixed parameters
fixed_model = "GFDL-ESM4"
fixed_directory = "/storage/group/pches/default/users/dcl5300/wbm_soilM_uc_2024_DATA/projections/eCONUS/out/LOCA2"
fixed_start_year = 2030
fixed_end_year = 2035
fixed_years = np.arange(fixed_start_year, fixed_end_year + 1)

target_lon = -88.8647
target_lat = 40.5478
location = 'Mclean_County'

# Variable parameters
possible_dirs = [
    "kge", "mae", "mse", "nse", "outer20rmse", "outer20ubrmse", "outer50rmse",
    "outer50ubrmse", "rmse", "taylor", "ubmae", "ubsme", "ubrmse"]

possible_dirs = ["SMAP", "VIC", "NOAH", "MOSAIC"]
possible_dirs = ["ssp245", "ssp370"]

base_pattern = "*r1i1p1f1_{}_MOSAIC_nse.nc"  # one placeholder for sub_dir
colors = ['y', 'r', 'b', 'g']

# If you want to vary something else (like models), you could define a list of models:
# possible_models = ["GFDL-ESM4", "AnotherModel", "YetAnotherModel"]
# and keep sub_dir fixed by removing the loop over possible_dirs
# and adding a loop over possible_models, etc.

# ---------------------------------------------------------------------------
# Data Processing and Plotting
# ---------------------------------------------------------------------------

plt.figure(figsize=(12, 4))
ci_range = []

for idx, sub_dir in enumerate(possible_dirs):
    pattern = base_pattern.format(sub_dir)
    files = glob.glob(os.path.join(fixed_directory, pattern))
    
    if not files:
        continue

    for f in files:
        model_name = os.path.basename(f).split("_")[0]

        if model_name == fixed_model:
            ds = xr.open_dataset(f)
            ds_small = ds.sel(time=slice(f"{fixed_start_year}-01-01", f"{fixed_end_year}-12-31"))

            subset = ds_small.sel(lon=target_lon, method='nearest').sel(lat=target_lat, method='nearest')
            ci_range.append(subset.soilMoist.values)
            yearly_data = subset.groupby("time.year").mean().soilMoist
            years = yearly_data["year"].values  # Extract years
            values = yearly_data.values  # Extract corresponding values
            
            degree = 1  # Linear regression
            coefficients = np.polyfit(years, values, degree)
            slope, intercept = coefficients
            
            # Calculate regression line
            regression_line_fixed = np.polyval(coefficients, fixed_years)
            
            plt.plot(fixed_years, regression_line_fixed, color =colors[idx])
            plt.scatter(years, values, label=f[-20:-1], color =colors[idx])
            ds.close()
            
            del ds_small

# Once we have gone through all files (for all sub_dirs for the fixed model)
# we can compute the uncertainty range
ci_95th = np.percentile(ci_range, 97.5, axis=0)
ci_5th = np.percentile(ci_range, 2.5, axis=0)

# Plot Confidence Interval with Daily Time
plt.fill_between(
    subset.time.values,
    ci_5th,
    ci_95th,
    color="b",
    alpha=0.1,
    label=f"95% Confidence Interval {fixed_model}"
)

plt.title(f"{fixed_start_year}-{fixed_end_year} Inter SSP Soil Moisture Uncertainty, {location} trend")
plt.xlabel("Time")
plt.ylabel("Soil Moisture")
plt.ylim(150, 450)
plt.legend(fontsize='small', loc='upper right')

# Uncomment to save the figure if desired
base_path = "/storage/home/cta5244/work/0_soilmoisture/0_uncertainity_figures/"
#plt.savefig(base_path + (f'1_soil_moist_model_{fixed_model}_{base_pattern[1:-1]}_{fixed_start_year}_{fixed_end_year}_{location}_trend.png').replace("{", "-").replace("}", "-"))

#plt.show()
```

## Integrated Economic Model

### Load Data, Select Relevant Values

```python
soil_moisture_data = pd.read_stata('soilMoistureData.dta')
soil_moisture_data['mrsoAprMay2'] = soil_moisture_data['mrsoAprMay'] ** 2
soil_moisture_data['mrsoJunJul2'] = soil_moisture_data['mrsoJunJul'] ** 2
soil_moisture_data['mrsoAugSep2'] = soil_moisture_data['mrsoAugSep'] ** 2

# Square of average daily soil moisture fraction
soil_moisture_data['smf2'] = soil_moisture_data['smf'] ** 2
soil_moisture_data['smf2'].attrs['label'] = "Square of mean daily soil moisture fraction"

# Square of average daily ET
soil_moisture_data['et2'] = soil_moisture_data['et'] ** 2
soil_moisture_data['et2'].attrs['label'] = "Square of mean daily evapotranspiration (mm)"

# Square of average daily soil moisture content (alternative interpolation)
soil_moisture_data['mrso_alt2'] = soil_moisture_data['mrso_alt'] ** 2
soil_moisture_data['mrso_alt2'].attrs['label'] = "Square of mean daily soil moisture content (mm)**, alternative interpolation"

# Probability of no compound stress
soil_moisture_data['shrDD29nl'] = soil_moisture_data['dd29smNl'] / (
    soil_moisture_data['dd29smLo'] + soil_moisture_data['dd29smHi'] + soil_moisture_data['dd29smNl']
)
soil_moisture_data['shrDD29nl'].attrs['label'] = "Share of DD29C at normal soil moisture"

# Probability of no compound or individual stress
soil_moisture_data['shrDD10_29nl'] = soil_moisture_data['dd10_29smNl'] / (
    soil_moisture_data['dd10_29smLo'] + soil_moisture_data['dd10_29smHi'] + soil_moisture_data['dd10_29smNl']
)
soil_moisture_data['shrDD10_29nl'].attrs['label'] = "Share of DD10-29C at normal soil moisture"

# ---- dummy variables
soil_moisture_data['d10'] = 0
soil_moisture_data.loc[soil_moisture_data['shrDD29nl'] < 0.1, 'd10'] = 1

soil_moisture_data['d90'] = 0
soil_moisture_data.loc[soil_moisture_data['shrDD29nl'] > 0.9, 'd90'] = 1

# ---- cleaning USDA data on irrigated area
soil_moisture_data['cornAreaIrrig'] = soil_moisture_data['cornArea'].where(
    (soil_moisture_data['cornAreaIrrig'].isna()) & (soil_moisture_data['longitude'] < -100),
    soil_moisture_data['cornAreaIrrig']
)
soil_moisture_data['cornAreaNonIrrig'] = soil_moisture_data['cornArea'].where(
    (soil_moisture_data['cornAreaNonIrrig'].isna()) & (soil_moisture_data['longitude'] > -100),
    soil_moisture_data['cornAreaNonIrrig']
)
soil_moisture_data['cornAreaIrrig'] = soil_moisture_data['cornAreaIrrig'].fillna(0)
soil_moisture_data['cornAreaNonIrrig'] = soil_moisture_data['cornAreaNonIrrig'].fillna(0)

soil_moisture_data['irAreaShr'] = soil_moisture_data['cornAreaIrrig'] / (
    soil_moisture_data['cornAreaIrrig'] + soil_moisture_data['cornAreaNonIrrig']
)
soil_moisture_data['irAreaShr'].attrs['label'] = "Share of irrigated corn area"

# Save the processed data to a new file
#soil_moisture_data.to_csv('processed_soil_moisture_data.csv', index=False)
```

### Crop Model

```python
# Load the data
data = pd.read_stata('soilMoistureData.dta')

# Set multi-index for panel data
data = data.set_index(['state', 'year'])  # Assuming 'state' and 'year' are panel identifiers

# Add time-related features for interaction terms
data['t'] = data.index.get_level_values('year')  # Time variable for interactions
data['t2'] = data['t'] ** 2  # Quadratic time variable

# Function to run fixed-effects regression
def run_fe_model(dependent, independent_vars, cluster_var, data):
    """Runs a fixed-effects regression with clustering."""
    exog = sm.add_constant(data[independent_vars])  # Add constant for the model
    model = PanelOLS(data[dependent], exog, entity_effects=True, check_rank=False)  # Fixed-effects model
    results = model.fit(cov_type='clustered', cluster_entity=True)  # Cluster by entity (state)
    return results

# Model (1-a): Basic model with precipitation
results_1a = run_fe_model(
    'logCornYield',
    ['dday10_29C', 'dday29C', 'prec', 'prec2', 't', 't2'],
    cluster_var='state',
    data=data
)

# Model (1-b): Includes soil moisture
results_1b = run_fe_model(
    'logCornYield',
    ['dday10_29C', 'dday29C', 'mrso', 'mrso2', 't', 't2'],
    cluster_var='state',
    data=data
)

# Model (1-c): Soil moisture thresholds
results_1c = run_fe_model(
    'logCornYield',
    ['dday10_29C', 'dday29C', 'nd0xSM_m_gt025', 'nd0xSM_m_lt025', 't', 't2'],
    cluster_var='state',
    data=data
)

# Model (1-d): Soil moisture deviations
results_1d = run_fe_model(
    'logCornYield',
    ['dday10_29C', 'dday29C', 'smdPos', 'smdNeg', 't', 't2'],
    cluster_var='state',
    data=data
)

# Model (2-a): Split heat index to individual and compound extremes
results_2a = run_fe_model(
    'logCornYield',
    ['dday10_29C', 'gdd29smxxx_75b', 'gdd29sm75b_25b', 'gdd29sm25b_25a',
     'gdd29sm25a_75a', 'gdd29sm75a_xxx', 'mrso', 'mrso2', 't', 't2'],
    cluster_var='state',
    data=data
)

# Model (2-b): Split soil moisture to individual and compound extremes
results_2b = run_fe_model(
    'logCornYield',
    ['dday10_29C', 'dday29C', 'mrNlTvg25_50', 'mrHiTvg25_50', 'mrLoTvg25_50',
     'mrLoTvg00_25', 'mrHiTvg00_25', 'mrNlTvg00_25', 't', 't2'],
    cluster_var='state',
    data=data
)

# Print results for one of the models
print(results_1a.summary)
```
